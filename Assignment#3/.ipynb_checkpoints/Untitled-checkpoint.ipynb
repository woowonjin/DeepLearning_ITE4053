```python
import time, random, sys

import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.models import Sequential

outfile = 'data.npz'

m = 10000    #training samples
n = 500     #test samples
k = 1000    #gradient updates
batch = 10000

x1_train=[]
x2_train=[]
y_train=[]

x1_test=[]
x2_test=[]
y_test=[]


if __name__ == '__main__':
#     x_train = np.load("x_train.npy")# (2,m)
#     x_train = x_train.transpose()
#     y_train = np.load("y_train.npy")  # (1, m)
#     x_test = np.load("x_test.npy")  # (n, 2)
#     y_test = np.load("y_test.npy")  # (1, n)
    
    X_ = np.load("x_train.npy").transpose()# (2,m)
    X_test = np.load("x_test.npy")
    Y_ = np.load("y_train.npy")
    Y_test = np.load("y_test.npy")
    
    print(X_.shape, X_test.shape, Y_.shape, Y_test.shape)

    input_layer = InputLayer(input_shape=(2,))
    hidden_layer = Dense(3, activation='sigmoid')
    output_layer = Dense(1, activation='sigmoid')

    model = Sequential([
    input_layer,
    hidden_layer,
    output_layer
    ])

    t1 = time.time()
    model.compile(
                optimizer="SGD",
                loss=tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error"),
                metrics=['accuracy'])

    model.fit(X_, Y_, epochs=k, batch_size= batch, verbose=2)
    t2 = time.time()
    model.evaluate(X_test,  Y_test, verbose=2)
```
